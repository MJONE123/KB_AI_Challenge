```python
import os

# 클래스 목록 및 라벨 할당
classes = ["bag", "earphone", "paper_bag", "phone", "umbrella", "wallet"]
class_to_label = {cls: i for i, cls in enumerate(classes)}

# 데이터셋 경로 설정
dataset_path = "C:/imgData"

# 각 클래스 폴더를 순회하며 라벨 파일 수정
for cls in classes:
    class_folder = os.path.join(dataset_path, cls)
    for split in ["test", "train", "valid"]:
        split_folder = os.path.join(class_folder, split, "labels")
        if not os.path.exists(split_folder):
            continue

        label_files = [f for f in os.listdir(split_folder) if f.endswith(".txt")]

        for label_file in label_files:
            label_path = os.path.join(split_folder, label_file)

            # 라벨 파일 읽기
            with open(label_path, 'r') as f:
                lines = f.readlines()

            # 기존 라벨을 새로운 라벨로 수정
            new_lines = []
            for line in lines:
                parts = line.strip().split()
                parts[0] = str(class_to_label[cls])  # 클래스 라벨 수정
                new_lines.append(" ".join(parts))

            # 수정된 라벨 파일 저장
            with open(label_path, 'w') as f:
                f.write("\n".join(new_lines))

print("라벨 수정 완료")
```
